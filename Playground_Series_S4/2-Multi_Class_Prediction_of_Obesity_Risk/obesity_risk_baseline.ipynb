```python
import pandas as pd
import numpy as np
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
df_train = pd.read_csv('data/train.csv')
df_original = pd.read_csv('data/ObesityDataSet.csv')
df_test = pd.read_csv('data/test.csv')

df_train = pd.concat((df_train, df_original), axis=0).drop(['id'], axis=1)
```

```python
train = TabularDataset(df_train)
test = TabularDataset(df_test)
```

```python
automl = TabularPredictor(label='NObeyesdad', problem_type='multiclass', eval_metric='accuracy')
automl.fit(train, presets='best_quality')
```

```python
automl.leaderboard()
```

```python
prediction = automl.predict(test)

data_submit = pd.read_csv('data/sample_submission.csv')
data_submit['NObeyesdad'] = prediction
data_submit[['id', 'NObeyesdad']].to_csv('data/baseline1.csv', index=False)
```
